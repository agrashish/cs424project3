```python
import pandas as pd
```

```python
c_movie_title = []
certificates = []
found = False

with open("certificates.list", 'r') as f:
    for x in range(14):
        next(f)
    for line in f:
        temp = list(filter(None, line.strip().split("	")))
        c_movie_title.append(temp[0])
        certificates.append(temp[-1])
        
certificate = pd.DataFrame(list(zip(c_movie_title, certificates)), columns=['Title', 'Certificate'])

#Dropping rows without USA certificate
index = certificate[certificate['Certificate'].str.contains('USA') != True].index
certificate = certificate.drop(index)
#Dropping rows that are TV
index = certificate[certificate['Certificate'].str.contains('USA:TV') == True].index
certificate = certificate.drop(index)
certificate = certificate[certificate['Certificate'].str.contains('USA:X') == False]
certificate = certificate[certificate['Certificate'].str.contains('USA:NC-17') == False]
#Dropping rows that have video game certificate
index = certificate[certificate['Certificate'].str.contains('E') == True].index
certificate = certificate.drop(index)
index = certificate[certificate['Certificate'].str.contains('T') == True].index
certificate = certificate.drop(index)
index = certificate[certificate['Certificate'].str.contains('C') == True].index
certificate = certificate.drop(index)
index = certificate[certificate['Certificate'].str.contains('M') == True].index
certificate = certificate.drop(index)
index = certificate[certificate['Certificate'].str.contains('Not Rated') == True].index
certificate = certificate.drop(index)
index = certificate[certificate['Certificate'].str.contains('Unrated') == True].index
certificate = certificate.drop(index)
index = certificate[certificate['Certificate'].str.contains('Not Rated') == True].index
certificate = certificate.drop(index)
certificate = certificate[certificate["Title"].str.contains("\(TV\)|\(V\)|\(VG\)") == False]
certificate = certificate[certificate['Title'].str.get(0) != "\""]
```

```python
# genres = pd.read_csv('genres.csv', encoding = 'latin1', usecols=['THE GENRES LIST'])
movie_title = []
genres_list = []
found = False

with open('genres.list') as genres:
    for g in genres:
        if g.strip() == '8: THE GENRES LIST':
            found = True
        elif found:
            if g.find('\t'):
                word = g.strip().split('\t')
                movie_title.append(word[0])
                genres_list.append(word[-1])
                
genre = pd.DataFrame(list(zip(movie_title, genres_list)), columns=['Title', 'Genres'])
genre = genre[2:]
genre = genre[genre["Title"].str.contains("\(TV\)|\(V\)|\(VG\)") == False]
genre = genre[genre['Title'].str.get(0) != "\""]
genre.head()

# genres to remove
remGens = ["Short", "Adult", "Reality-tv", "Reality-TV", "Talk-Show", "Game-Show", "News"]
counts = genre["Genres"].value_counts()
counts = counts[counts < 100]
for item in counts.index:
    if item not in remGens:
        remGens.append(item)
remGens

remString = ""
for gen in remGens:
    remString += gen
    remString += "|"
remString = remString[:-1]
remString

remMovies = genre[genre['Genres'].str.contains(remString) == True]
remMovies.head()

genre = genre[genre["Title"].isin(remMovies["Title"]) == False]
```

```python
genre
```

```python
movie = []
key = []
found = False;

with open('keywords.csv') as keys:
    for line in keys:
        if line.strip() == '8: THE KEYWORDS LIST':
            found = True
        elif found:
            if(line.find('\t')):
                word = line.strip().split('\t')
                movie.append(word[0])
                key.append(word[-1])
            
keyword = pd.DataFrame(list(zip(movie, key)), columns=['Title', 'Keyword'])

keyword = keyword[keyword["Title"].str.contains("\(TV\)|\(V\)|\(VG\)") == False]
keyword = keyword[keyword['Title'].str.get(0) != "\""]
keyword = keyword.iloc[2:]
keyword
```

```python
values = keyword['Keyword'].value_counts()
key_list = keyword['Keyword'].value_counts().index.tolist()
value_list = keyword['Keyword'].value_counts().values.tolist()
keys_to_drop = []
keys = keyword.copy()
for i in range(len(values)):
    if (value_list[i] < 20):
         keys_to_drop.append(key_list[i])
            
keys = keyword[keyword['Keyword'].isin(keys_to_drop) == False]
keys
```

```python
movie_list = []
movie_year = []
found = False

with open('movies.csv') as movies:
    for movie in movies:
        if movie.strip() == 'MOVIES LIST':
            found = True
        elif found:
            if movie.find('\t'):
                movie_seperate = movie.strip().split('\t')
                movie_list.append(movie_seperate[0])
                movie_year.append(movie_seperate[-1])
                
movies = pd.DataFrame(list(zip(movie_list,movie_year)), columns=['Title', 'Year'])

#Dropping rows that have ???? has a year
index = movies[movies['Year'].str.contains('\?\?\?\?') == True].index
movies = movies.drop(index)
movies = movies[movies["Title"].str.contains("\(TV\)|\(V\)|\(VG\)") == False]
movies = movies[movies['Title'].str.get(0) != "\""]
movies = movies.iloc[2:]
movies = movies.iloc[:-1]
```

```python
movies
```

```python
ratings_list = []
Distribution = []
Votes = []
Rank = []
Title = []
found = False

with open('ratings.csv', 'r') as ratings:
    for rating in ratings:
        if rating.strip() == 'New  Distribution  Votes  Rank  Title':
            found = True
        elif rating.strip() == 'REPORT FORMAT':
            break
        elif found:
            ratings_seperate = rating.strip().split('  ')
            if len(ratings_seperate) > 3:
                Distribution.append(ratings_seperate[0].strip())
                Votes.append(ratings_seperate[1].strip())
                Rank.append(ratings_seperate[-2].strip())
                Title.append(ratings_seperate[-1].strip())
                
ratings = pd.DataFrame(list(zip(Title,Rank)), columns=['Title', 'Rating'])

ratings = ratings[ratings["Title"].str.contains("\(TV\)|\(V\)|\(VG\)") == False]
ratings = ratings[ratings['Title'].str.get(0) != "\""]
```

```python
ratings
```

```python
r_movie_title = []
release_dates = []
found = False

with open('release-dates.csv') as release:
    for r in release:
        if r.strip() == 'RELEASE DATES LIST':
            found = True
        elif found:
            if r.find('\t'):
                r_seperate = r.strip().split('\t')
                r_movie_title.append(r_seperate[0])
                if ('(' in r_seperate[-1]) == False :
                    release_dates.append(r_seperate[-1].split(':')[-1])
                else:
                    release_dates.append(r_seperate[-2].split(':')[-1])
        
release = pd.DataFrame(list(zip(r_movie_title, release_dates)), columns=['Title', 'Release Date'])

#Removing just the year
release = release[release["Release Date"].str.contains(" ") == True]
#Removing rows for the release date internet
index = release[release['Release Date'].str.contains('\(internet\)') == True].index
release = release.drop(index)
#Removing rows for the release date re-release
index = release[release['Release Date'].str.contains('\(re-release\)') == True].index
release = release.drop(index)
#Removing rows for the release date blue-ray
index = release[release['Release Date'].str.contains('\(Blu-ray premiere\)') == True].index
release = release.drop(index)
release = release[release["Title"].str.contains("\(TV\)|\(V\)|\(VG\)") == False]
release = release[release['Title'].str.get(0) != "\""]
release = release.iloc[1:]
release = release.iloc[:-1]
release.drop_duplicates(subset ="Title", keep = False, inplace = True) 
```

```python
release
```

```python
run_movie_title = []
running_times = []

found = False

with open('running-times.csv', 'r') as running:
    for run in running:
        if run.strip() == 'RUNNING TIMES LIST':
            found = True
        elif found:
            run_seperate = run.strip().split('\t')
            run_movie_title.append(run_seperate[0])
            running_times.append(run_seperate[-1])

runningTime = pd.DataFrame(list(zip(run_movie_title, running_times)), columns=['Title', 'Running Time'])
runningTime = runningTime.iloc[1:]
runningTime = runningTime.iloc[:-1]
runningTime = runningTime[runningTime["Title"].str.contains("\(TV\)|\(V\)|\(VG\)") == False]
runningTime = runningTime[runningTime['Title'].str.get(0) != "\""]
runningTime = runningTime[runningTime['Running Time'].str.isnumeric() == True]
runningTime['Running Time'] = runningTime['Running Time'].astype(int)

runningTime = runningTime[runningTime['Running Time'] > 59]
```

```python
runningTime
```

```python
frames = [certificate, genre, keyword, movies, ratings, release, runningTime]

joined = certificate.join(genre.set_index('Title'), on='Title')
joined = joined.join(keyword.set_index('Title'), on='Title')
joined = joined.join(movies.set_index('Title'), on='Title')
joined = joined.join(ratings.set_index('Title'), on='Title')
joined = joined.join(release.set_index('Title'), on='Title')
joined = joined.join(runningTime.set_index('Title'), on='Title')
joined = joined.dropna()

joined.to_csv('movies_data.csv', index=False)
```
